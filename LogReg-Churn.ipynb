```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# train a logistic regression model on the training set
from sklearn.linear_model import LogisticRegression


# instantiate the model
logreg = LogisticRegression(solver='liblinear', random_state=0)


# fit the model
logreg.fit(X_train, y_train)
```

```python
y_pred_test = logreg.predict_proba(X_test)

y_pred_test
#for each record it will give two columns, probability of churning and probablity of not churning
```

```python
y_pred_test_class = logreg.predict(X_test)
#Ytrain will have only dependent variable,Xtest will have independent variable
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(y_test, y_pred_test_class)
confusion_matrix
```

```python
X_test = pd.DataFrame(X_test)
churn_pred = pd.DataFrame(y_pred_test)
churn_pred.columns = ['No_Churn','Churn']
churn_pred.head()
X_test
```

```python
test_data = pd.concat ([X_test,y_test],axis = 1)
test_data
```

```python
test_data.reset_index(drop=True, inplace=True)
churn_pred.reset_index(drop=True, inplace=True)
#adding predicted column


df_pred = pd.concat( [test_data, churn_pred], axis=1) 
df_pred.to_csv('C:/Users/nisha/OneDrive/Desktop/IV Sem/ML/churn/Output/predicted_churn.csv')
```

```python
print('\nTrue Positives(TP) = ', confusion_matrix[0,0])

print('\nTrue Negatives(TN) = ', confusion_matrix[1,1])

print('\nFalse Positives(FP) = ', confusion_matrix[0,1])

print('\nFalse Negatives(FN) = ', confusion_matrix[1,0])
```

A false positive is when we determines something is true when it is actually false (also called a type I error)
A false negative is saying something is false when it is actually true (also called a type II error)

precision = TP/ (TP+FP)

recall = TP/ (TP+FN)

specificity = TN/(TN+FP)

sensitivity = recall
